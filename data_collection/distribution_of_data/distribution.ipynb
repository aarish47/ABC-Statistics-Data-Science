# **Distribution of Data**

```python
# import the libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import scipy 
from scipy import stats
```

```python
# load the dataset
df = sns.load_dataset("tips")
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='total_bill', kde=True)
```

```python
# load the other dataset
df_01 = sns.load_dataset("titanic")
df_01.head()
```

```python
# creating a histplot
sns.histplot(data=df_01, x='age', kde=True, bins=25)
```

```python
sns.histplot(data=df_01, x='age', kde=True, bins=25, hue='class')
```

```python
sns.boxplot(data=df_01, x='age', hue='class')
```
